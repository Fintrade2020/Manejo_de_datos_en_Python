$$

## Estandarizaci√≥n estad√≠stica.

(valor actual - media)/desviacion estandar


$$
\frac{valoractual - media}{desviacion}
$$

```python
df["length"] = (df["length"] - df["length"].mean())/df["length"].std()
df["length"].head()
df[["width", "height", "length"]].head()
```

## Data binning.

Se usa para agrupar valores dependiendo de el valor de la observacion.

```python
df["price"] = df["price"].astype("float") #se pasa price a float.
bins = np.linspace(min(df["price"]),max(df["price"]),4) #se crea el vector con los rangos.
etiquetas = ["bajo","medio","alto"] #Etiquetas para la nueva columna.
df["binning"] = pd.cut(df["price"], bins, labels= etiquetas,
                       include_lowest=True) #Se crea la columna nueva con las etiquetas de los rangos.
df.head(10)
```

```python
df["price"] = df["price"].astype("float")
bins = [0,15000,20000,50000] #Se establecen rangos manualmente.
etiquetas = ["bajo","medio","alto"]
df["binning"] = pd.cut(df["price"], bins, labels= etiquetas, include_lowest=True)
df.head(10)
```

## Crear funciones en python.

$$
\frac{"valoractual" - media}{desviacion}
$$

```python
def estandarizacion(v):
  media= v.astype("float").mean()
  desviacion = v.std()

  return((v-media)/desviacion)
```
